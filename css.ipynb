1. You are building a landing page for a small bakery in your city. The page looks fine
on your laptop but when you open it on your phone, the text is too small and the images go
outside the screen. Which CSS technique will you use to fix this and how will you start?

   - Will add tag meta name="viewport" content="width=device-width, initial-scale=1.0"
   - Also i will use media queries if in case we need have to change font-size for mobile, we can increase it or decrease it.
   - for images we will give img {
  max-width: 100%;
  height: auto;
} and container max-width

2. You added a navbar at the top of your website. The client wants the navbar to
always stay visible at the top, even when the user scrolls down. Which CSS property will you
use and what value will you set?

   - Position fixed with top value zero(0) and width 100%

3. You wrote CSS for a button — first you added a class .btn with blue background,
then added an id #submitBtn with red background to the same button. The button is showing
red instead of blue. Why is this happening?

    - Id(#btn) selectors will have  higher priority so it will overide the value given on .btn class.

4. You are building a product card for an online store. You want three cards to sit side
by side on desktop, but stack one below the other on mobile. Which layout system will you use
and how will you make it responsive?

   - For desktop use
   .container {
  display: flex;
  gap: 20px;
}

.card {
  flex: 1; // gives you equal space for all element
}
for mobile use
@media (max-width: 768px) {
  .container {
    flex-direction: column; // break all cards into below to each other
  }
}

8. What is the difference between Flexbox and CSS Grid? When should you use each?

   - Flexbox is used for one-dimensional layouts (row or column).
      CSS grid is used for 2-d layouts

9. What are the three ways to add CSS to an HTML page? Which one is preferred for
real projects and why?

  -  Inline , Internal and External
External is preferred most bacause
* Keeps HTML and CSS separate
* Allows the same stylesheet to be reused across multiple pages.
* Makes the code easier to manage and update.
* Browsers can cache the CSS file, which can improve loading performance on subsequent visits.

10. What is a media query in CSS? Write a small example that changes the
background color when the screen is smaller than 600px.

 - Media query helps in making responsive design, it helps to make sure tablet, mobile and desktop screen work fine without breaking the layout.
 @media (max:width: 600px){
  .element{
  background:red;
  }
 }
